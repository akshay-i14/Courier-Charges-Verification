### Processing the data

```python
print('SKUs duplicates data:')
display(SKU_Master[SKU_Master.duplicated()])
print('Pincode Zones duplicates data:')
display(pincode_zones[pincode_zones.duplicated()])
```

```python
# Droping duplicates
SKU_Master.drop_duplicates(inplace=True)
pincode_zones.drop_duplicates(inplace=True)
```

```python
# converting weight from gram(G) to kilogram(KG)
SKU_Master['Weight (g)'] = SKU_Master['Weight (g)'] /1000 
SKU_Master.rename(columns = {'Weight (g)':'Weight(kg)'}, inplace = True)
```

```python
display(SKU_Master.head())
```

```python
# Getting proper order_report based on SKU 
order_report = pd.merge(order_report,SKU_Master, how = 'inner', on = 'SKU')
```

```python
display(order_report)
```

```python
# Calculating total weight by multiplying Weight and Quantity
order_report['Total weight as per X (KG)'] = order_report['Order Qty'] * order_report['Weight(kg)']
```

```python
display(order_report.head())
```

```python
# Creating order_report_2 by droping some field
order_report_2 = order_report.drop(['SKU','Order Qty','Weight(kg)'],axis=1)
```

```python
# Total weight as per order id by using groupby()
order_report_2 = order_report_2.groupby(by = ['ExternOrderNo'])[['Total weight as per X (KG)']].sum().reset_index()
```

```python
display(order_report_2)
```

```python
# creating function for weight slab
def pointfive_slab(data):
    if floor(data) + 0.5 == data:
        output = data
    elif round(data/10,1) * 10 >= data:
        output = round(data/10,1) * 10
    else:
        output = round(data/10,1) * 10 + 0.5
        
    return output  
```

```python
# Calculating weight as per weight slab by using function
order_report_2['Weight slab as per X (KG)'] = order_report_2['Total weight as per X (KG)'].apply(pointfive_slab)
```

```python
display(order_report_2)
```

```python
# Creating detailed invoice by merging invoice and order_report_2 on order id
invoice_2 = pd.merge(invoice, order_report_2, how='inner', left_on ='Order ID', right_on = 'ExternOrderNo' )
invoice_2.drop('ExternOrderNo', axis=1, inplace=True)
invoice_2.head()
```

```python
# Checking unique value of warehouse pincode
print('Warehouse pincode in X company dataset:',pincode_zones['Warehouse Pincode'].unique())
print('Warehouse pincode in Courier company dataset:',invoice_2['Warehouse Pincode'].unique())
```

```python
# Merging invoice_2 and pincode zones
invoice_2 = pd.merge(invoice_2, pincode_zones, how='inner', left_on = ['Warehouse Pincode','Customer Pincode'], right_on = ['Warehouse Pincode','Customer Pincode'])
```

```python
display(invoice_2)
```

```python
# Applying slab function to get weight slabs charge by courier company
invoice_2['Weight slab charged by Courier Company (KG)'] = invoice_2['Total weight as per Courier Company (KG)'].apply(pointfive_slab)
```

```python
# Extracting zone
column = rates.columns
temp_set = set()

for i in column:
    temp_set.add(i[4:5])

temp_lst = list(temp_set)
temp_lst.sort()
temp_lst
```

```python
rates_1 = pd.DataFrame({'Zones':temp_lst})
rates_1
```

```python
# Function for extracting rates for corresponding to each 'zone' and 'ship'
def fwd_fixed_charges(zone):
    fwd_fixed = rates[f'fwd_{zone}_fixed']
    return fwd_fixed
    
def fwd_additional_charges(zone):
    fwd_additional = rates[f'fwd_{zone}_additional']
    return fwd_additional

def rto_fixed_charges(zone):
    rto_fixed = rates[f'rto_{zone}_fixed']
    return rto_fixed
    
def rto_additional_charges(zone):    
    rto_additional = rates[f'rto_{zone}_additional']
    return rto_additional
```

```python
# Applying function
rates_1['fwd_fixed'] = rates_1['Zones'].apply(fwd_fixed_charges)
rates_1['fwd_additional'] = rates_1['Zones'].apply(fwd_additional_charges)
rates_1['rto_fixed'] = rates_1['Zones'].apply(rto_fixed_charges)
rates_1['rto_additional'] = rates_1['Zones'].apply(rto_additional_charges)
```

```python
rates_1
```

```python
print('Shipment Charges:',invoice_2['Type of Shipment'].unique())
```

```python
# Function for calculating shipping charges based on zone and shipping type
def shipping_charges(zone,shipment, weight):
    lst = []
    
    for i in range(len(zone)):
        charges = 0
        charges = float(rates_1['fwd_fixed'][rates_1['Zones']==zone[i]].values)
        updated_weight = weight[i] - 0.5
        
        if updated_weight > 0:
            count = updated_weight / 0.5
            charges = charges + float(rates_1['fwd_additional'][rates_1['Zones']==zone[i]].values) * count
        
        if shipment[i] == 'Forward and RTO charges':
            charges = charges + float(rates_1['rto_fixed'][rates_1['Zones']==zone[i]].values)
            if updated_weight > 0:
                charges = charges + float(rates_1['rto_additional'][rates_1['Zones']==zone[i]].values) * count
        
        charges = round(charges,1)
        lst.append(charges)

                
    return lst
```

```python
# Applying shipping charge function to calculate Shipping charge as per company X
lst = shipping_charges(invoice_2['Delivery Zone as per X'], invoice_2['Type of Shipment'], invoice_2['Weight slab as per X (KG)'])
```

```python
# Dataframe for charges
charges = pd.DataFrame({'Expected Charge as per X (Rs.)':lst})
```

```python
charges.head()
```

```python
# Concating charge as per X and Courier company
charges = pd.concat([charges,invoice_2['Charges Billed by Courier Company (Rs.)']],axis=1)
```

```python
# Shipping charge difference between company x and courier company
charges['Difference Between Expected Charges and Billed Charges (Rs.)'] = charges['Expected Charge as per X (Rs.)'] - charges['Charges Billed by Courier Company (Rs.)']
```

```python
display(charges)
```

```python
# Converting zone to upper case as per desired output
invoice_2['Delivery Zone as per X']= invoice_2['Delivery Zone as per X'].str.upper()
invoice_2['Delivery Zone charged by Courier Company']= invoice_2['Delivery Zone charged by Courier Company'].str.upper()
```

```python
# final invoice by rearraning columns of invoice_2 and concating invoice_2 with charges 
final_invoice = pd.concat([invoice_2[['Order ID','AWB Number','Total weight as per X (KG)','Weight slab as per X (KG)','Total weight as per Courier Company (KG)','Weight slab charged by Courier Company (KG)','Delivery Zone as per X','Delivery Zone charged by Courier Company']],charges],axis=1)
```

```python
display(final_invoice)
```

```python
# Summary calculation (Charges difference)
final_charges = {}
```

```python
# Total orders where X has been correctly charged 
final_charges['Total Orders - Correctly Charged'] = [charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0].shape[0]
                                                     , round(charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0]['Expected Charge as per X (Rs.)'].sum(),1)]
```

```python
# Total Orders where X has been overcharged 
final_charges['Total Orders - Over Charged'] = [charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0].shape[0]
                                                , round(charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum(),1)]
```

```python
# Total Orders where X has been undercharged
final_charges['Total Orders - Under Charged'] = [charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0].shape[0]
                                                 , round(charges[charges['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum(),1)]
```
